```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import gridspec
```

```python
data=pd.read_csv('creditcard.csv')
data
```

```python
fraud=len(data[data['Class']==1])
valid=len(data[data['Class']==0])
print(fraud)
print(valid)
```

```python
correlation =  data.corr()
plotting = plt.figure(figsize=(7,7))
sns.heatmap(correlation,vmax= .10,square=True)
plt.show()
```

```python
x=data.drop(['Class'],axis=1)
y=data['Class']
x_data=x.values
y_data=y.values
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x_data,y_data,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

rand=RandomForestClassifier()
rand.fit(x_train,y_train)
prediction=rand.predict(x_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score
from sklearn.metrics import precision_score,recall_score
from sklearn.metrics import f1_score,matthews_corrcoef
from sklearn.metrics import confusion_matrix
```

```python
n_outliers=fraud
n_errors=(prediction!=y_test).sum()
print("This model uses Randomforest Classificaation")

acc=accuracy_score(y_test,prediction)
print("The Accuracy Score is {}".format(acc))

prec=precision_score(y_test,prediction)
print(f"The Precision Score is {prec}")

rec=recall_score(y_test,prediction)
print(f"The Recall Score is {rec}")

f1=f1_score(y_test,prediction)
print(f"The f1 Score is {f1}")

MCC=matthews_corrcoef(y_test,prediction)
print(f"The Matthews_corrcoef is {MCC}")
```

```python
conf_matrix=confusion_matrix(y_test,prediction)
conf_matrix
```
